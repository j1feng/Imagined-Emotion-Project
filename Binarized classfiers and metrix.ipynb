```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
```

Number of predictions that are correct:

```python
metrics.accuracy_score(y_test, preds)
```

|Outcome of Prediction|Definition|True Class|
|---|---|---|
|True positive (TP)|the model correctly predicts the positive class.|P|
|False negative (FN)|the model incorrectly predicts the negative class.|P|
|True negative (TN)|the model correctly predicts the negative class.|N|
|False positive (FP)|the model incorrectly predicts the positive class.|N|


Find the counts of TN / FP / FN / TP

```python
metrics.confusion_matrix(y_test, preds)
```

Sensitivity/Recall: What is the proportion of actual (rare) positives were correctly identified?


Precision: What proportion of positive identifications were actually correct? (when you decide to identify a passenger as a terrorist, how often are you right?)

Since we have all the positive as positive emotions

In this case: `sensitivity` is what is the proportion of actual positive emotions that are correct identified?
sensitivity = how good it is to detect postive emotion

`Precision` is when we label an event as positive emotion, what is the proportion that we actually get it correct?

```python
#proportion of actual positive emotions that are correct identified
metrics.recall_score(y_test, preds)
```

```python
#proportion of actual negative emotions that are correct identified
metrics.recall_score(y_test, preds, pos_label=0)
```

```python
#proportion of predicted positive emotions that are actually positive emotions
metrics.precision_score(y_test, preds)
```

```python
#proportion of predicted positive emotions that are actually negative emotions
1 - metrics.precision_score(y_test, preds)
```
